### Numbers

Python's four number types are integers,floats, complex numbers and booleans:

* *Integers*: 1, -3, 42,355, 8888, -7777777 (integers aren't limited in seze except by available memory)
* *Floats*: 3.0, 31e12, -6e-4
* *Complex numbers*: 3+2j, -4-2j
* *Booleans*: True, False

You can manipulate them by using the arithmetic operators: + (addition),-(subtraction), *(multiplication), / (division), ** (exponentation), and % (modulus).

```python
5/2 # floating-point result with normal division
```

```python
5//2 # integer result with truncation when divided usin //
```

```python
3**3 # exponentitation
```

```python
2.0e-8 # Scientific notation give back a float
```

### Strings

```python
# Indicates strings through de use of double quotes.
x = "Hello,world"
x
```

```python
# You can use single quotes instead of double quotes. The following two lines do the same thing:
x = "Hello,World"
print(x)
x = 'Hello,World'
print(x)
```

```python
# The only difference is that you don't need to backslash " characters in single-quoted strings or 
# ' characters in double-quoted strings:
x = "Don't need a backslash"
x = 'Cant\'t get by without backslash'
x = "Backslash your \" character!"
x = 'You can leave the " alone'
```

```python
# Python offers triple-quoted strings, which let you do this and include single and double quotes without backslashes
x = """ Starting and ending a string with triple " character 
permits embedded newlines, and the use of " and
' without
backslashes
"""
print(x)
```

### Casting variables
There may be times when you want to specify a type on to a variable.

Casting in python is therefore done using functions:
* $\texttt{int()}$ - constructs an integer number from an integer, a float (by removing all decimals), or a string (providing the string represents a whole number)

* $\texttt{float()}$- constructs a float number from an integer, a float  or a string (providing the string represents a float or an integer) 

* $\texttt{str()}$ - constructs a string from a wide variety of data types, including strings, integer  and float 


```python
# Integer
x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3
print(x,y,z)
```

```python
# Float
x = float(1)     # x will be 1.0
y = float(2.8)   # y will be 2.8
z = float("3")   # z will be 3.0
w = float("4.2") # w will be 4.2
print(x,y,z,w)
```

```python
# String
x = str("s1") # x will be 's1'
y = str(2)    # y will be '2'
z = str(3.0)  # z will be '3.0'
print(x,y,z)
```
