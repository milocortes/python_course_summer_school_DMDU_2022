## Entropía

Dado un conjunto de eventos con probabilidad $P_i$ correspondiente a una información particular para cada evento, entonces el promedio ponderado de la información (en bits) está dada por la entropía de Shannon.

Dada una distribución de probabilidad $p_1,p_2,\dots,p_N$, la entropía de Shannon, $H$, es igual a :

\begin{equation}
    H(p_1,p_2,\dots,p_N) = -\sum_{i=1}^N p_i \log_2 p_i
\end{equation}


Pensemos la economía como un proceso que, de forma aleatoria, asigna trabajadores a la división social del trabajo.

Las ocupaciones en las que es asignado un individuo es representada como una variable aleatoria $s$.

Los autores consideran que la cantidad promedio de información que es requerida para describir en qué ocupación termina el individuo está dada por la entropía de Shannon.


El sistema económico constituye un proceso con sectores interconectados *¿cómo incorporamos la idea de entropía?*

Leontief consideraba que el sistema económico estaba constituido de múltiples insumos y productos para distintos bienes y servicios.

Distintos sectores utilizan y producen insumos. 

Existe una interdependencia entre distintos sectores. 

Una fracción del producto del sector $i$ es requerida para la producción en el sector $j$.

Supongamos que existen $n$ sectores interconectados que reproducen el sistema económico ($S$).

Cada sector necesita trabajo para producir. 

Asumamos que el trabajo directo requerido por cada sector está dado por $li$. 

Ahora consideremos que una unidad  del trabajo social total es asignado a un sector inicial ($s0$) de forma aleatoria.

Entonces, la probabilidad que una unidad de trabajo sea asignada a un sector está dada por la siguiente expresión, la cual da el número promedio de bits requeridos para codificar el sector inicial.

La entropía de un sector inicial aleatorio $s_0$ está dado por :

\begin{equation}
    H(s_0) = - \sum_{i = 1}^N  \pi_i \log_2 \pi_i \geq 0,
\end{equation}

donde 

\begin{equation}
    \pi_i = \dfrac{l_i}{\sum_j l_j} 
\end{equation}

es la probabilidad que una unidad de trabajo sea asignada a producir en el sector $i$ ($l_i$ es el trabajo directo requerido en cada sector).

¿Qué significa esto en términos económicos? imaginemos una economía preindustrial en la que la mayoría de trabajadores se encuentran en la agricultura y sólo unos cuantos en sectores no agrícola.  En este caso, la entropía es baja.

En la medida que la industrialización traslada trabajadores hacia sectores no agrícolas, la entropía de la economía se incrementa. La máxima entropía se alcanzaría cuando todos los sectores requieren la misma cantidad de trabajo.

Los autores formulan el proceso de producción de Leontief como una cadena (modelo) de Markov. Estos modelos capturan las transiciones probabilísticas en un conjunto finito de estados que cumple la propiedad de Markov:

\begin{equation}
    P(X_{t+1} = s| X_1 = s_1, \dots ,X_t = s_t) = P(X_{t+1} = s |X_t = s_t)
\end{equation}

Una fracción de la producción en un sector $s0$ es el insumo de un sector $s1$. Podemos verlo como una transición aleatoria $s_0 \rightarrow s_1$. 

Considerando un modelo de Markov, una unidad de producto del sector $s_0$ se convierte en insumo del sector $j$ con una probabilidad $p_{ij}$, la cual corresponde a la fracción del producto del sector $i$ requerido para la producción en el sector $j$.

De manera que se considera que la transición aleatoria $s_0 \rightarrow s_1$ ocurre con probabilidad Pr$(s_i=j | s_0 = i) = p_{ij}$



 **¿Cómo calculamos la entropía de esta transición?** 
 
 La entropía del sector $s_1$, subsecuente al sector $s_0$, está dada por la expresión:
 
 \begin{equation}
     H(s_1 | s_0 = i) = - \sum_{j_=1}^N p_{ij} \log_2 p_{ij}
 \end{equation}

El significado económico es el siguiente. 

Imaginemos una economía industrial anterior a la electrificación y que el sector $i$ es electricidad generada por centrales eléctricas que usan carbón. 

Inicialmente pocos sectores utilizan electricidad como insumo, de manera que la entropía $H(s_1|s_0=i)$ es baja. 

En la medida que la electricidad es incorporada a los procesos productivos, esta entropía se incrementa.

El número de bits requerido para codificar $s_1$, dado un estado inicial aleatorio $s_0$ se obtiene con la siguiente expresión:

\begin{equation}
        H(s_1|s_0) = \sum_{i=1}^n \pi_i H(s_1|s_0 = i) 
\end{equation}
Conforme la economía se diversifica y más sectores requieren una combinación grande de insumos, se incrementa la entropía condicional $H(s_1|s_0)$.

Los autores extienden este enfoque a través de subsecuentes sectores en el proceso de producción. 

Pensemos una secuencia de $k+1$ variables aleatorias, $s_0 \rightarrow s_1 \rightarrow \dots \rightarrow s_k$, que representa los requerimientos de producción de $k$-ésimo orden.

Leontief consideró el proceso de producción como un *flujo circular* de requerimientos, de manera que la secuencia $s_0 \rightarrow s_1 \rightarrow \dots \rightarrow s_k$ es una representación del proceso cuando $k$ tiende al infinito.

Los autores indican que al conceptualizar de esta forma el proceso productivo, es posible encontrar un límite finito a la entropía condicional, la cual indicaría el número de bits requerido para codificar, en el límite, a $s_k$. 

Así, la entropía del sistema es igual a la expresión:

\begin{equation}
    H(S) = - \sum_{i=1}^n\sum_{j=1}^n\pi_i^*p_{ij}\log_2 p_{ij}
\end{equation}

donde  los elementos $\pi_i^*$ están dados por el eigenvector  $\mathbf{\pi^*} = \mathbf{P\pi^*}$ con eigenvalor igual a 1. Es decir, es una solución no trivial a $(\mathbf{I}-\mathbf{P})\mathbf{\pi^*} = \mathbf{0}$

La entropía está acotada por

\begin{equation}
    0 \leq H(S) \leq \log_2n
\end{equation}

y logra su valor máximo cuando la transición entre sectores es la misma.

Con dicha expresión podemos calcular la entropía con las matrices anuales de insumo-producto que el INEGI publicó recientemente: https://www.inegi.org.mx/temas/mip/


```python
import pandas as pd
import numpy as np
import math
import scipy
from scipy.sparse.linalg import eigs
import requests
import matplotlib.pyplot as plt


import os

# Cambiamos al directorio de descargas
os.chdir("../descargas")
# Descargamos las MIP a precios constantes de 2013
url = "https://www.inegi.org.mx/contenidos/investigacion/mcsm/tabulados/"
mi_inegi = ["mip_pcn_ixi_t_{}.xlsx".format(x) for x in range(2003,2019)]

for mip in mi_inegi:
    print("Descargando {}".format(mip))
    r=requests.get(url+mip, allow_redirects=True)
    open(mip,'wb').write(r.content)
```

```python

# Definimos función que calcula la entropía
def compute_entropy(archivo):
    ## Cargamos la matriz que se encuentra en formato .xlsx
    mip = pd.read_excel(archivo, index_col=None, na_values=['-'],header =0, skiprows=4, nrows=258,usecols = "B:IY")
    ## Convertimos el data frame en la matriz Z
    Z = mip.to_numpy()
    ### Nos quedamos con los sectores "básicos".
    ## Se excluyen a los sectores con filas iguales a cero
    delete_sectors = []

    for i in range(Z.shape[0]):
        suma = sum(Z[i,])
        if suma==0:
            delete_sectors.append(i)

    for i in range(len(delete_sectors)):
        if i ==0:
            Z=np.delete(Z, delete_sectors[i], axis=0)
            Z=np.delete(Z,  delete_sectors[i], axis=1)
        else:
            Z=np.delete(Z, delete_sectors[i]-i, axis=0)
            Z=np.delete(Z,  delete_sectors[i]-i, axis=1)
    ## Construimos la matriz de probabilidades o de transiciones
    P = np.zeros((Z.shape[0],Z.shape[0]))
    suma = 0
    for i in range(Z.shape[0]):
        P[i,] = Z[i,]/sum(Z[i,])

    ##  Calculamos el vector de estacionaridad-probabilidad
    values, vectors = scipy.sparse.linalg.eigs(P, k=1, sigma=1)
    pi_vector = vectors/sum(vectors)
    ## Se calcula la entropía en bits
    H = 0
    for i in range(P.shape[0]):
        for j in range(P.shape[0]):
            if P[i][j]>0:
                H-= pi_vector[i]*P[i][j]*math.log2(P[i][j])
    print("Porcentaje del nivel máximo de entropía {}".format(H[0].real/math.log2(P.shape[0])))
    return H[0].real
```

```python

### Calculamos la entropía para las MIP de 2003 a 2020
print("%--------------------------------------------------%")
print("               Calculando entropía")
print("%--------------------------------------------------%")

H_mip=[]

for mip in mi_inegi:
    print(mip)
    H_mip.append(compute_entropy(mip))

### Obtenemos el dato del índice de complejidad económica

print("%--------------------------------------------------%")
print("Obtenemos el dato del índice de complejidad económica")
print("%--------------------------------------------------%")

complex_index = pd.read_csv("Country Complexity Rankings 1995 - 2018.csv")
mexico= complex_index[complex_index['Country']=='Mexico']
mexico_eci = np.around(mexico[['ECI {}'.format(anio) for anio in range(2003,2019)] ].to_numpy(),2)[0].tolist()
```

```python

### Graficamos las series
tiempo_line = [x for x in range(2003,2019)]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Año')
ax1.set_ylabel('Entropía', color=color)
ax1.plot(tiempo_line, H_mip, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel("Economic Complexity Index (ECI)", color=color)
ax2.plot(tiempo_line, mexico_eci, color=color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()
plt.savefig('../output/entropia_vs_eci.jpg')
```

Comparamos los resultados con el Economic Complexity Index (ECI) del Growth Lab de Harvard. En dicho índice los países aumentan el ECI al aumentar el número y complejidad de los productos que exportan.

Para el periodo 2003-2011, ambos indicadores muestran una caída sostenida. El ECI presenta un incremento para el periodo 2012-2016, mientras que la entropía muestra una mejoría para 2013-2015. Ambos indicadores muestran una caída para 2017 y 2018.

La entropía muestra una caída sostenida para todo el periodo. Los autores sugieren que caídas así pueden explicarse por el declive de la industria manufacturera.
